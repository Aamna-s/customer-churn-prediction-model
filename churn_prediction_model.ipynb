# Pre processing


```python
# col_features = {col:np.mean(features[col].dropna()) for col in features.columns if features[col].isna().any()}
# print(col_features)
# for col in col_features.keys():
#     features[col].fillna(col_features[col], inplace=True)
```

```python
# One Hot Encoding  (Encoding catageroy technique , we do this to make easy for model to predict the numerical value/flags)
features = pd.get_dummies(features, columns=["Geography", "Gender"])
features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Feature Scaling (scale the data [0,1] so that every param can contribute equally rather than dominancy/deficiency)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# fit --> mean,std + tranform--> apply
#That means you told the scaler:
#Hey, forget training data. Learn new averages from test data.”
#This is bad because your model should not look at test data during training (data leakage) bcz somehow it learn any patterns from train data which might effect the learning. So, no need to fit from test data. train fit will auto apply.
X_test = scaler.fit_transform(X_test)
X_train[0]
```

```python
def handle_outliers(data):
  numeric_cols = data.select_dtypes(include=[np.number]).columns
  for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    InterquartileRange = Q3 - Q1
    lower_bound = Q1 - 1.5 * InterquartileRange
    upper_bound = Q3 + 1.5 * InterquartileRange
    data[col] = np.where(data[col] > upper_bound, upper_bound, np.where(data[col] < lower_bound, lower_bound, data[col]))
  return data
```

```python
X_train = pd.DataFrame(X_train, columns=features.columns)
X_test = pd.DataFrame(X_test, columns=features.columns)

X_train = handle_outliers(X_train)
X_test = handle_outliers(X_test)
```

# Model Training


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
print(f"numpy == {np.__version__}")
```

```python
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
lr_predictions= lr.predict(X_test)
accuracy = accuracy_score(y_test, lr_predictions)
accuracy
```

```python
def evaluate(model,X_train, X_test, y_train, y_test , filename):
    model.fit(X_train,y_train)
    y_test_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_test_pred)
    print(f"{model.__class__.__name__} Accuracy { accuracy} ")
    print(f"{model.__class__.__name__} Classification Report { classification_report(y_test, y_test_pred)} ")
    print("---------------------------")
    with open (filename,'wb') as file:
        joblib.dump(model, file)
    print(f"model save das {filename}")
```

```python
rf = RandomForestClassifier(random_state=42)
evaluate(rf, X_train, X_test, y_train, y_test, "rf_model.pkl")

dt = DecisionTreeClassifier(random_state=42)
evaluate(dt, X_train, X_test, y_train, y_test, "dt_model.pkl")

kn = KNeighborsClassifier()
evaluate(kn, X_train, X_test, y_train, y_test, "kn_model.pkl")

svc = SVC(random_state=42)
evaluate(svc, X_train, X_test, y_train, y_test, "svc_model.pkl")

gnb = GaussianNB()
evaluate(gnb, X_train, X_test, y_train, y_test, "gnb_model.pkl")

xgboost = xgb.XGBClassifier(random_state=42)
evaluate(xgboost, X_train, X_test, y_train, y_test, "xgboost_model.pkl")
```

# Feature Engineering

```python
features['AgeGroup'] = pd.cut(features['Age'], bins=[0,30,45,60,100], labels=['Young', 'MiddleAge', 'Senior','Elderly'])
features
```

```python
features['TenureAgeRatio'] = df['Tenure']/df['Age']
features
```

```python
features = pd.get_dummies(features, drop_first=True)
features.columns
```

```python
# Now after adding feature test again on Model we pick before
X_train, X_test, y_train, y_test = train_test_split(features, df['Exited'], test_size=0.2, random_state=42)
xgb1 = xgb.XGBClassifier(random_state=42)
evaluate(xgb1,X_train,X_test, y_train, y_test, 'xgb_model.pkl')
# Now the accurancy is improve and also precesioon, f1 score but the recall we focus on reduce so not beneficial. but it poves that feature engineer changes params a lot like accuracy increased by 10
```
